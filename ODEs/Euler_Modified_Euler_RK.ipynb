```python
import numpy as np
import matplotlib.pyplot as plt

# Problem: y' = sin(x),  y(0) = 1  ->  y(x) = 2 - cos(x)
def f(x, y):        # y doesn't appear, but we keep the signature
    return np.sin(x)

def y_true(x):
    return 2 - np.cos(x)

# Integrators (fixed step)
def euler(f, x0, x1, y0, h):
    xs = np.arange(x0, x1 + 1e-12, h)
    ys = np.empty_like(xs)
    ys[0] = y0
    for n in range(len(xs) - 1):
        ys[n+1] = ys[n] + h * f(xs[n], ys[n])
    return xs, ys

def modified_euler(f, x0, x1, y0, h):
    xs = np.arange(x0, x1 + 1e-12, h)
    ys = np.empty_like(xs)
    ys[0] = y0
    for n in range(len(xs) - 1):
        k1 = f(xs[n], ys[n])
        y_pred = ys[n] + h*k1
        k2 = f(xs[n] + h, y_pred)
        ys[n+1] = ys[n] + 0.5*h*(k1 + k2)
    return xs, ys

def rk4(f, x0, x1, y0, h):
    xs = np.arange(x0, x1 + 1e-12, h)
    ys = np.empty_like(xs)
    ys[0] = y0
    for n in range(len(xs) - 1):
        x, y = xs[n], ys[n]
        k1 = f(x, y)
        k2 = f(x + 0.5*h, y + 0.5*h*k1)
        k3 = f(x + 0.5*h, y + 0.5*h*k2)
        k4 = f(x + h,     y + h*k3)
        ys[n+1] = y + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)
    return xs, ys

# ------------------- play with these -------------------
x0, x1, y0 = 0.0, 5.0, 1.0
hs = [0.5, 0.2, 0.05]   # try e.g. [1.0, 0.2, 0.05, 0.01]
# -------------------------------------------------------

for h in hs:
    xe, ye = euler(f, x0, x1, y0, h)
    xh, yh = modified_euler(f, x0, x1, y0, h)
    xr, yr = rk4(f, x0, x1, y0, h)
    xt = np.linspace(x0, x1, 1000)
    yt = y_true(xt)

    # errors (just to see the trend)
    def max_err(x, y): return np.max(np.abs(y - y_true(x)))
    print(f"h = {h:>6}:  Euler {max_err(xe, ye):.3e} | Modified_euler_err {max_err(xh, yh):.3e} | RK4 {max_err(xr, yr):.3e}")

    # plot
    plt.figure()
    plt.plot(xe, ye, "o-", label=f"Euler (h={h})")
    plt.plot(xh, yh, "s-", label=f"Modified_euler  (h={h})")
    plt.plot(xr, yr, "^-", label=f"RK4   (h={h})")
    plt.plot(xt, yt, label="Analytical")
    plt.xlabel("x"); plt.ylabel("y")
    plt.title(f"y' = sin(x), y(0)=1   (step h = {h})")
    plt.legend()
    plt.show()
```
